```python
wo_clean = [word.strip().lower() for word in wo_list]
```

```python
wo_unique=list(set(wo_clean))
```

```python
wo_unique.sort(key=lambda s:len(s))
```

```python
from collections import defaultdict
dec=defaultdict(list)


for i in range(23):
    for y in wo_unique[:]:
        if i == len(y):
              dec[i].append(y)
```
